# Introduction
In this project, I used K-Means Clustering, a popular unsupervised machine learning technique, to analyze the Red Wine Quality dataset.
K-Means is a method that groups similar data points into clusters based on their features.
Since the true labels (wine quality scores) are not necessarily clear-cut categories, clustering helps to explore hidden structures in the data.

Real-world usage of K-Means includes customer segmentation, market basket analysis, document classification, and image compression.
Here, I focused on clustering wines based on their chemical properties to see if distinct groups naturally form.

# Step 6: Using the Elbow Method to find optimal K

```python
# Apply Elbow Method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

# Plot Elbow
plt.figure(figsize=(8,6))
plt.plot(range(1,11), wcss, marker='o')
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.grid(True)
plt.show()
```

# Elbow Method Interpretation
In this step, I used the Elbow Method to help choose the optimal number of clusters for K-Means clustering.

The plot shows the WCSS (Within-Cluster Sum of Squares) for different numbers of clusters, ranging from 1 to 10.
WCSS measures how close the data points are to their assigned cluster centers — lower WCSS values are better.

At first, WCSS drops sharply as the number of clusters increases, which makes sense because more clusters can explain the data better.
However, after a certain point, adding more clusters doesn’t significantly reduce WCSS — the curve flattens.

When I look at the curve, I can see a noticeable "elbow" around k = 3.
This is where the improvement in WCSS starts to slow down.

👉 Therefore, based on the Elbow Method, I selected k = 3 as the best number of clusters.
Choosing k=3 balances between having enough clusters to capture meaningful structure and not overcomplicating the model.

# Step 7: Train KMeans Model (Density and Alcohol)

```python
# Set K=3
kmeans = KMeans(n_clusters=3, random_state=0)
clusters = kmeans.fit_predict(X)
centroids = kmeans.cluster_centers_

# Plot clusters with centroids
fig, ax = plt.subplots(figsize=(10,8))
scatter = ax.scatter(X[:,0], X[:,1], c=clusters, cmap='viridis', edgecolor='k')
ax.scatter(centroids[:,0], centroids[:,1], c='red', marker='X', s=300, label='Centroids')
ax.set_xlabel('Density')
ax.set_ylabel('Alcohol')
ax.set_title('K-Means Clustering (Density vs Alcohol)')
plt.legend()
plt.grid(True)
plt.show()
```

In this plot, I visualized the clustering of the wines based on density and alcohol.

Each dot represents one wine sample.

Colors indicate cluster membership.

The red crosses (X) mark the centroids of each cluster.

From this scatter plot, I noticed that alcohol content plays an important role in separating the clusters. Wines with higher alcohol tend to group together, while those with lower alcohol form another group.

The clusters are fairly well-separated, which suggests that alcohol and density are useful features for clustering wines.

# Step 8: Calculate and Print Silhouette Score

```python
# Calculate Silhouette Score
silhouette_avg = silhouette_score(X, clusters)
print(f"Silhouette Score for K=3 clusters: {silhouette_avg:.3f}")
```

To validate the clustering quality, I calculated the Silhouette Score:

Silhouette measures how close each point in one cluster is to points in neighboring clusters.

Higher values are better (maximum is 1).

📈 Result:

For k=3 clusters, the Silhouette Score was 0.605.

This suggests that the clustering structure is quite good and distinct.


# Step 9: Clustering with 3 Features (Density, pH, Alcohol)

```python
# Select features: density, pH, alcohol
X = df.iloc[:, [7, 8, 10]].values

# Standardize
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply Elbow Method again
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=0)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(8,6))
plt.plot(range(1,11), wcss, marker='o')
plt.title('The Elbow Method (Density, pH, Alcohol)')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.grid(True)
plt.show()
```

To enrich the analysis, I added pH as a third feature alongside Density and Alcohol.

After scaling the features (standardization), I re-applied the Elbow Method.

# Plot Explanation:

Again, the elbow appeared clearly around k=2 or 3.

This suggests the data could be grouped into 2 or 3 natural clusters even in 3D space

# Step 10: Final KMeans and 3D Visualization

```python
# K=2
kmeans = KMeans(n_clusters=2, random_state=0)
clusters = kmeans.fit_predict(X_scaled)

# 3D Scatter plot
fig = plt.figure(figsize=(12,10))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X_scaled[:,0], X_scaled[:,1], X_scaled[:,2], c=clusters, cmap='viridis', edgecolor='k')
ax.set_xlabel('Density')
ax.set_ylabel('pH')
ax.set_zlabel('Alcohol')
ax.set_title('KMeans Clustering (3 Features)')
plt.show()
```

Finally, I trained a KMeans model with k=2 clusters (as the simpler choice) using three features: Density, pH, and Alcohol.

📊 Plot Explanation:

This is a 3D scatter plot showing the clusters based on the three features.

Points are colored by their cluster assignment.

Clear separation between clusters can be seen, especially along the density and alcohol axes.
