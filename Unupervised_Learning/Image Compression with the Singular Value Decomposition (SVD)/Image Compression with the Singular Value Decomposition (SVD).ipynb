In this project, I explored image compression using Singular Value Decomposition (SVD).
The idea is to efficiently compress an image by approximating it with fewer singular values while still retaining as much of the visual quality as possible.
This method is particularly useful in real-world applications where reducing image storage size without major quality loss is critical.

# Step 1: Import Libraries
In this step, I imported the necessary libraries:

numpy for numerical operations,

matplotlib for visualization,


The try-except block ensures that if ipywidgets isn't installed, the notebook will still run without interactive mode.

```python

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.image import imread

```

# Step 2: Load an RGB Image
Here, I loaded the image of Peynir (saved as peynir.jpg) and checked its pixel value range.
If pixel values were not between 0 and 1, I normalized them accordingly.
This normalization ensures that all further computations stay consistent and avoids color distortion

```python

# Make sure you have an image file in the same directory, e.g., 'dog.jpg'
img = imread('peynir.jpg')  # Replace with your file name!

# Check if pixel values are between 0-255 and normalize to 0-1 if necessary
if img.max() > 1.0:
    img = img / 255.0

print(f"Image shape: {img.shape}")
plt.imshow(img)
plt.title('Original Image')
plt.axis('off')
plt.show()
```

# Step 3: Separate Channels

```python

R = img[:, :, 0]
G = img[:, :, 1]
B = img[:, :, 2]
```

# Step 4: Apply SVD to each color channel
Using numpy.linalg.svd, I decomposed each color channel into:

U (left singular vectors),

S (singular values),

V (right singular vectors).

These matrices form the basis of reconstructing the compressed images at different quality levels

```python

Ur, Sr, Vr = np.linalg.svd(R, full_matrices=False)
Ug, Sg, Vg = np.linalg.svd(G, full_matrices=False)
Ub, Sb, Vb = np.linalg.svd(B, full_matrices=False)

print(f"Shapes: Ur={Ur.shape}, Sr={Sr.shape}, Vr={Vr.shape}")
```

```python
# üé® To have fun: Visualize R, G, B channels separately

fig, axs = plt.subplots(1, 3, figsize=(15, 5))

axs[0].imshow(R, cmap='Reds')
axs[0].set_title('Red Channel')
axs[0].axis('off')

axs[1].imshow(G, cmap='Greens')
axs[1].set_title('Green Channel')
axs[1].axis('off')

axs[2].imshow(B, cmap='Blues')
axs[2].set_title('Blue Channel')
axs[2].axis('off')

plt.suptitle('Separate Visualization of RGB Channels', fontsize=16)
plt.show()
```

# Step 5: Define a function to reconstruct the compressed image

I created a reusable reconstruct_image(k) function:

It keeps only the first k singular values for each color channel.

Reconstructs a compressed approximation of the original image.

Ensures the resulting image values stay within the valid pixel range [0,1].

This function is critical for efficiently generating compressed versions of the image at various ranks!

```python

def reconstruct_image(k):
    R_comp = np.dot(Ur[:, :k], np.dot(np.diag(Sr[:k]), Vr[:k, :]))
    G_comp = np.dot(Ug[:, :k], np.dot(np.diag(Sg[:k]), Vg[:k, :]))
    B_comp = np.dot(Ub[:, :k], np.dot(np.diag(Sb[:k]), Vb[:k, :]))
    
    # Stack the channels back
    compressed_img = np.stack([R_comp, G_comp, B_comp], axis=2)
    
    # Clip values to stay within [0,1] after reconstruction
    compressed_img = np.clip(compressed_img, 0, 1)
    
    return compressed_img
```

# Step 6: Visualize Compressed Images at Different Ranks
I plotted the compressed images for several k values:

k = 5, 20, 50, 100, 200, 300

 Observation:
As the rank k increases, the image quality improves ‚Äî but so does the size!
With a low k, Peynir's features are blurred, while a higher k restores more detail.



```python

k_values = [5, 20, 50, 100, 200, 300]

plt.figure(figsize=(15, 10))
for idx, k in enumerate(k_values, 1):
    plt.subplot(2, 3, idx)
    img_comp = reconstruct_image(k)
    plt.imshow(img_comp)
    plt.title(f"Compressed Image\nk={k}")
    plt.axis('off')

plt.tight_layout()
plt.show()
```

# Step 7: Calculate and print compression ratios

```python

m, n = R.shape

for k in k_values:
    compression_ratio = (k * (m + n + 1)) / (m * n)
    print(f"For k = {k}, Compression Ratio: {compression_ratio:.4f}")
```
