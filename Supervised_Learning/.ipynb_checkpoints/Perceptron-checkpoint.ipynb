# Perceptron Learning
In this project, I implemented a Perceptron model from scratch to predict customer purchase behavior based on two features: Annual Income and Spending Score.
The process involves understanding basic supervised learning concepts, preparing the dataset, visualizing feature relationships, training a Perceptron classifier, and evaluating its performance.

# Key concepts covered:

Data preprocessing
Decision boundary visualization
Learning curve analysis
Error trends over epochs

# Strengths:
Simple and easy to implement
Fast convergence for linearly separable data
Useful for binary classification tasks

# Limitations:
Cannot solve non-linearly separable problems (like XOR)
Sensitive to feature scaling and learning rate selection

# Learning Process:
The model iteratively adjusts its parameters based on prediction errors until convergence is reached or a maximum number of epochs is exceeded.

# ðŸ“š Step 4: Build and Train the Perceptron Model

```python


# Prepare input X and output y
X = df[['Annual Income (k$)', 'Spending Score (1-100)']].values
y = df['Purchase'].values

# Convert labels to -1 and 1 for Perceptron
y = np.where(y == 0, -1, 1)

# Perceptron class
class Perceptron:
    def __init__(self, eta=0.01, epochs=1000):
        self.eta = eta
        self.epochs = epochs

    def fit(self, X, y):
        self.w_ = np.random.randn(X.shape[1] + 1)  # Random init
        self.errors_ = []

        for _ in range(self.epochs):
            errors = 0
            for xi, target in zip(X, y):
                update = self.eta * (target - self.predict(xi))
                self.w_[1:] += update * xi
                self.w_[0] += update
                errors += int(update != 0.0)
            self.errors_.append(errors)
        return self

    def net_input(self, X):
        return np.dot(X, self.w_[1:]) + self.w_[0]

    def predict(self, X):
        return np.where(self.net_input(X) >= 0.0, 1, -1)


# Instantiate and train the Perceptron
ppn = Perceptron(eta=0.01, epochs=1000)
ppn.fit(X, y)
```

# ðŸ“š Step 5: Plot Decision Boundary

```python
# Decision Boundary
plt.figure(figsize=(10,6))
plot_decision_regions(X, y, clf=ppn)
plt.title('Decision Boundary: Income vs Spending Score')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```

# Plot Analysis: Decision Boundary
This plot shows the learned decision boundary between customers predicted to purchase (+1) and not purchase (-1).

Customers above the boundary (higher spending scores relative to income) are predicted to purchase.

Customers below the boundary are predicted to not purchase.

The boundary is linear because the Perceptron can only draw straight lines to separate classes.

# ðŸ“š Step 7: Final Error Trend Check

```python
# Error trend over epochs
plt.figure(figsize=(10,5))
plt.plot(range(1, len(ppn.errors_) + 1), ppn.errors_, marker='o')
plt.xlabel('Epochs')
plt.ylabel('Number of misclassifications')
plt.title('Misclassifications over epochs')
plt.show()
```
