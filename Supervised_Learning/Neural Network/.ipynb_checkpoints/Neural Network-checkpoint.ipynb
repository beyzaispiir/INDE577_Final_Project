# Neural Network
In this project, I built a custom neural network using TensorFlow to classify images from the CIFAR-10 dataset. CIFAR-10 contains 60,000 32x32 color images in 10 different classes such as airplane, automobile, ship, and horse.
The goal was to train a model that can predict the correct class given a new image.

Strengths: Easy-to-use TensorFlow API, fast model building.

Limitations: CIFAR-10 images are small and slightly blurry, making the classification harder. Also, using a fully connected (dense) network without convolution layers limits performance.

# 📚 Step 1: Import Libraries and Load Data
I imported TensorFlow, NumPy, Matplotlib, and Seaborn.
I loaded the CIFAR-10 dataset and normalized pixel values to [0,1].

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix

# Load CIFAR-10 dataset
cifar10 = tf.keras.datasets.cifar10
(X_train, y_train), (X_test, y_test) = cifar10.load_data()

# Normalize pixel values to [0, 1]
X_train = X_train / 255.0
X_test = X_test / 255.0

print(f"Train shape: {X_train.shape}, Test shape: {X_test.shape}")

# 📊 Plot sample images
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer', 
               'dog', 'frog', 'horse', 'ship', 'truck']

plt.figure(figsize=(10,10))
for i in range(16):
    plt.subplot(4,4,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(X_train[i])
    plt.xlabel(class_names[y_train[i][0]])

plt.suptitle('Sample Images from CIFAR-10 Dataset')
plt.tight_layout()
plt.show()
```

I plotted 16 random images from the training set to see what kind of pictures we are working with.

Result:

Images displayed with correct labels (frog, truck, automobile, etc.)

Images are small and slightly blurry because CIFAR-10 images are naturally low resolution.

Analysis:

Some classes (like truck vs automobile) look similar, which might make it harder for the model to distinguish.

Normalization helped standardize pixel values but doesn't "sharpen" images.

I plotted both Loss Curve and Accuracy Curve.

Result:

Training loss slowly decreased.

Validation loss fluctuated, meaning model is not learning perfectly.

Training and validation accuracy curves stayed close but noisy.

Analysis:

High variance in validation accuracy indicates overfitting or underfitting.

Probably needs a CNN to better capture spatial patterns in images.

# 🧪 Step 4: Testing the Model

```python
# Evaluate
test_loss, test_accuracy = model.evaluate(X_test, y_test, verbose=2)
print(f"\nTest Accuracy: {test_accuracy:.4f}")
```

# 🧩 Step 5: Confusion Matrix

```python
# Predictions
y_pred_probs = model.predict(X_test)
y_pred = np.argmax(y_pred_probs, axis=1)

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(10,8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
plt.title('Confusion Matrix')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.show()

# Classification Report
print(classification_report(y_test, y_pred, target_names=class_names))
```

Best F1 score: ship (0.61)

Worst F1 score: cat (0.35)

Overall test accuracy: 48%

Analysis:

Classes like cat and bird were hard for the model to classify.

Ship and automobile classes had relatively higher accuracy.
