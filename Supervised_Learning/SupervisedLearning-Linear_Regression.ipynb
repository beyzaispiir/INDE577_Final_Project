```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the dataset (Area vs. Price)
X = np.array([50, 60, 70, 80, 90, 100, 110, 120, 130, 140])  # House area (in m²)
y = np.array([165, 195, 206, 237, 258, 299, 325, 387, 413, 442])  # House price (in thousand $)
```

```python
# Visualize the dataset
plt.scatter(X, y, color='blue')
plt.xlabel('House Area (m²)')
plt.ylabel('House Price (thousand $)')
plt.title('House Area vs Price')
plt.show()
```

```python
# Linear Regression Model (from scratch)
class LinearRegressionScratch:
    def __init__(self):
        self.m = None
        self.b = None

    def fit(self, X, y):
        n = len(X)
        numerator = (n * np.sum(X * y)) - (np.sum(X) * np.sum(y))
        denominator = (n * np.sum(X * X)) - (np.sum(X) ** 2)
        self.m = numerator / denominator
        self.b = (np.sum(y) - self.m * np.sum(X)) / n

    def predict(self, X):
        return self.m * X + self.b
```

```python
# Train the model
model = LinearRegressionScratch()
model.fit(X, y)

# Make predictions
y_pred = model.predict(X)
```

```python
# Plot the original data and the model's prediction
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, y_pred, color='red', label='Predicted Line')
plt.xlabel('House Area (m²)')
plt.ylabel('House Price (thousand $)')
plt.title('House Area vs Price (Model)')
plt.legend()
plt.show()
```

```python
# Print model parameters
print(f"Slope (m): {model.m}")
print(f"Intercept (b): {model.b}")
```

```python
# Predict the price of a new house with 160 m² area
new_area = 160
predicted_price = model.predict(new_area)
print(f"Predicted price for a house with {new_area} m² area: {predicted_price} thousand dollars")
```
